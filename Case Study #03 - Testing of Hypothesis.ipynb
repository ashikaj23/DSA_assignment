# Case Study on Testing of Hypothesis

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_rel
from scipy import stats
```

## A company started to invest in digital marketing as a new way of their product promotions.For that they collected data and decided to carry out a study on it.

```python
sales_data = pd.read_csv('Sales_add.csv')
```

```python
sales_data
```

```python
# Rename the Columns 
```

```python
sales_data.columns=['Month', 'Region', 'Manager', 'Sales_before_digital', 'Sales_After_digital']
sales_data
```

## ● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing

   Let u1 = mean of Sales_After_digital and u2 = mean of Sales_before_digital

- H0 (Null hypothesis) : (u1<=u2) There is no significant increase in sales after stepping into digital marketing. 
- Ha (Alternative hypothesis) : (u1>u2)  There is a significant increase in sales after stepping into digital marketing 

 Here we need to perform two sample paired t-test

```python
ttest_rel(sales_data['Sales_After_digital'],sales_data['Sales_before_digital'], alternative='greater')
```

 Here we could see that the p-value = 3.168e-11 is significantly smaller than 0.05.

 - Hence we reject the Null Hypothesis H0.

Therefore, There is a significant increase in sales after stepping into digital marketing.

## ● The company needs to check whether there is any dependency between the features “Region” and “Manager”

Inorder to check whether there is any dependency between the features “Region” and “Manager”, we need to use Chi-Sqaure test of independence

```python
sales_data
```

```python
rm=pd.crosstab(sales_data.Region,sales_data.Manager)
rm
```

Now the could find the crosstab frequencies of the two groups.

Let,

- H0 (Null Hypothesis): There is no dependency between Region and Manager
- Ha (Alternative Hypothesis): There is a dependency between Region and Manager

For Chi-square testing use the below syntax

```python
stats.chi2_contingency(rm)
```

The first value (3.050) is the Chi-square value, followed by the p-value (0.54939), then comes the degrees of freedom (4), and lastly it outputs the expected frequencies as an array.

Here we could see that the p-value = 0.549 is significantly larger than 0.05.

- Hence we fail to reject the Null Hypothesis H0.

Therefore, There is no dependency between Region and Manager.
