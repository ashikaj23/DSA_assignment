## ● Check whether all the zones generate the same amount of sales.

One sided ANOVA testing

- H0 (Null Hypothesis) : All the zones generate the same amount of sales
- Ha (Altervative Hypothesis) : All the zones generate different amount of sales

F statistics = Variance due to difference between groups/Error variance

```python
fvalue ,pvalue=stat.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
print('p-value = ',pvalue)
```

```python
print('f-value = ',fvalue)
```

- We reject the Null Hypothesis H0.
- As P-value = 0.0011827601694503335 is significantly smaller than 0.05.
- Therefore, All the zones generate different amount of sales
