```python
plt.figure(figsize=(6,15))

plt.subplot(4,1,1)
plt.scatter(Iris_data['Classification'],Iris_data['SL'],s=12, c='red',marker='^')
plt.title('Plot of Classification VS SL',fontsize=16)
plt.xlabel('Classification')
plt.ylabel('SL')

plt.subplot(4,1,2)
plt.scatter(Iris_data['Classification'],Iris_data['SW'],s=12, c='Blue',marker='^')
plt.title('Plot of Classification VS SW',fontsize=16)
plt.xlabel('Classification')
plt.ylabel('SW')

plt.subplot(4,1,3)
plt.scatter(Iris_data['Classification'],Iris_data['PL'],s=12, c='Green',marker='^')
plt.title('Plot of Classification VS PL',fontsize=16)
plt.xlabel('Classification')
plt.ylabel('PL')

plt.subplot(4,1,4)
plt.scatter(Iris_data['Classification'],Iris_data['PW'],s=12, c='Violet',marker='^')
plt.title('Plot of Classification VS PW',fontsize=16)
plt.xlabel('Classification')
plt.ylabel('PW')

plt.tight_layout()
```
